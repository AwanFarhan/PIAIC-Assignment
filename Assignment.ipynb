```python
statement = "my Name is Farhan and I like people with emotional an intellectual depth, people who speak with passion from their inner twisted mind"


#capitalize()
f = statement.capitalize()
print(f)
print("\n")
del f

#casefold()
f = statement.casefold()
print(f)
print("\n")
del f


#center()
name="Farhan"
f = name.center(20, "0")
print(f)
print("\n")


#count()
people="people"
f = statement.count(people)
print(f)
print("\n")
del f


#encode()
x="pythön!"
f = x.encode()
print(f)
print("\n")
del f
del x


#endswith()
x = "I am a  proud pakistani."
f = x.endswith("proud")
print(f)
f = x.endswith(".")
print(f)
print("\n")
del f
del x


#expandtabs()
x = "I\tam\ta\tproud\tpakistani"
f = x.expandtabs(16)
print(f)
print("\n")
del f
del x


#find()
f = statement.find("usa")
print(f)
#f = statement.index("usa")
#print(f)
print("\n")
del f


#isalpha()
x = "I am a  proud pakistani"
f = x.isalpha()
print(f)
x = "Iamaproudpakistani"
f = x.isalpha()
print(f)
print("\n")
del f
del x


#format()
x = "I am a  proud pakistani 1234"
f= "I am {} proud pakistani {}".format("a", 12345)
print(f)
f= "I am {0} proud pakistani {1}".format("a", 12345)
print(f)
f= "I am {a} proud pakistani {num}".format(a="a", num=12345)
print(f)
f= "I am {} proud pakistani {}".format("a", 12345)
print(f)
print("\n")
del f
del x


#isalnum()
x = "Iamaproudpakistani"
print(x.isalnum())
x = "I am a  proud pakistani"
print(x.isalnum())
del x
print("\n")


#isdecimal()
x = "28212"
print(x.isdecimal())
x = "32ladk3"
print(x.isdecimal())
x = "Mo3 nicaG el l22er"
print(x.isdecimal())
del x
print("\n")


#isdigit()
x = "28212"
print(x.isdigit())
x = "Mo3 nicaG el l22er"
print(x.isdigit())
del x
print("\n")


#isidentifier()
x = 'Python'
print(x.isidentifier())
x = 'Py thon'
print(x.isidentifier())
x = '22Python'
print(x.isidentifier())
x = ''
print(x.isidentifier())
del x
print("\n")


#islower()
s = 'this is good'
print(s.islower())
s = 'th!s is a1so g00d'
print(s.islower())
s = 'this is Not good'
print(s.islower())
del s
print("\n")


#isnumeric()
s = '1242323'
print(s.isnumeric())
#s = '²3455'
s = '\u00B23455'
print(s.isnumeric())
# s = '½'
s = '\u00BD'
print(s.isnumeric())
s = '1242323'
s='python12'
print(s.isnumeric())
del s
print("\n")


#isprintable()
s = 'Space is a printable'
print(s)
print(s.isprintable())
s = '\nNew Line is printable'
print(s)
print(s.isprintable())
s = ''
print('\nEmpty string printable?', s.isprintable())
del s
print("\n")


#isspace()
s = '   \t'
print(s.isspace())
s = ' a '
print(s.isspace())
s = ''
print(s.isspace())
del s
print("\n")


#istitile()
s = 'Python Is Good.'
print(s.istitle())
s = 'Python is good'
print(s.istitle())
s = 'This Is @ Symbol.'
print(s.istitle())
s = '99 Is A Number'
print(s.istitle())
s = 'PYTHON'
print(s.istitle())
del s
print("\n")


#isupper()
string = "THIS IS GOOD!"
print(string.isupper());
string = "THIS IS ALSO G00D!"
print(string.isupper());
string = "THIS IS not GOOD!"
print(string.isupper());
print("\n")


#filter()
alphabets = ['a', 'b', 'd', 'e', 'i', 'j', 'o']
def filterVowels(alphabet):
    vowels = ['a', 'e', 'i', 'o', 'u']
    if(alphabet in vowels):
        return True
    else:
        return False
filteredVowels = filter(filterVowels, alphabets)
print('The filtered vowels are:')
for vowel in filteredVowels:
    print(vowel)
print("\n")


#float()
print(float(10))
print(float(11.22))
print(float("-13.33"))
print(float("     -24.45\n"))
print("\n")

#input()
name = input()
rint('The entered name is:', name)
print("\n")


#int()
print("int(123) is:", int(123))
print("int(123.23) is:", int(123.23))
print("int('123') is:", int('123'))
print("\n")


#next()
vowels = ['a', 'e', 'i', 'o', 'u']
vowelsIter = iter(vowels)
print(next(vowelsIter))
print(next(vowelsIter))
print(next(vowelsIter))
print(next(vowelsIter))
print(next(vowelsIter))
print("\n")


#len()
testList = []
print(testList, 'length is', len(testList))
testList = [1, 2, 3]
print(testList, 'length is', len(testList))
testTuple = (1, 2, 3)
print(testTuple, 'length is', len(testTuple))
testRange = range(1, 10)
print('Length of', testRange, 'is', len(testRange))
print("\n")


#max()
print('Maximum is:', max(1, 3, 2, 5, 4))
num = [1, 3, 2, 8, 5, 10, 6]
print('Maximum is:', max(num))
print("\n")


#min()
print('Minimum is:', min(1, 3, 2, 5, 4))
num = [3, 2, 8, 5, 10, 6]
print('Minimum is:', min(num))
print("\n")


#map()
def calculateSquare(n):
  return n*n
numbers = (1, 2, 3, 4)
result = map(calculateSquare, numbers)
print(result)
numbersSquare = set(result)
print(numbersSquare)
print("\n")


#ord()
print(ord('5'))
print(ord('A'))
print(ord('$'))
print("\n")


#reversed()
seqString = 'Python'
print(list(reversed(seqString)))
seqTuple = ('P', 'y', 't', 'h', 'o', 'n')
print(list(reversed(seqTuple)))
seqRange = range(5, 9)
print(list(reversed(seqRange)))
seqList = [1, 2, 4, 3, 5]
print(list(reversed(seqList)))
print("\n")


#sum()
numbers = [2.5, 3, 4, -5]
numbersSum = sum(numbers)
print(numbersSum)
numbersSum = sum(numbers, 10)
print(numbersSum)
print("\n")

#zip()
numberList = [1, 2, 3]
strList = ['one', 'two', 'three']
result = zip()
resultList = list(result)
print(resultList)
result = zip(numberList, strList)
resultSet = set(result)
print(resultSet)
print("\n")
```
